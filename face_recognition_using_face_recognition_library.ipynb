```python
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread('facezuckerberg1.jpg')
plt.imshow(img)
plt.show()
```

```python
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.title('image')
plt.imshow(img)
plt.show()
```

```python
img_loc =face_recognition.face_locations(img)[0]
print(img_loc)
encode_img = face_recognition.face_encodings(img)[0]
output = cv.rectangle(img, (img_loc[3], img_loc[0]), (img_loc[1], img_loc[2]), (255, 0, 255), 2)

plt.imshow(output)
plt.show()
```

```python
imgtest = cv.imread('facezuckerberg.jpg')
imgtest = cv.cvtColor(imgtest, cv.COLOR_BGR2RGB)
imgtest_loc = face_recognition.face_locations(imgtest)[0]
imgtest_encodings = face_recognition.face_encodings(imgtest)[0]


outputtest = cv.rectangle(imgtest, (imgtest_loc[3], imgtest_loc[0]), (imgtest_loc[1], imgtest_loc[2]), (255, 0, 255), 2)

plt.imshow(outputtest)
plt.show()
```

compare the image with the other image

```python
result  = face_recognition.compare_faces([encode_img], imgtest_encodings)
resultDistance = face_recognition.face_distance([encode_img], imgtest_encodings)
resultoutput = cv.putText(imgtest, f'{result} {round(resultDistance[0],2)} ', (50, 50), cv.FONT_HERSHEY_COMPLEX, 1, (255, 0, 255), 3)
plt.imshow(resultoutput)
plt.show()
print(resultDistance)
```
