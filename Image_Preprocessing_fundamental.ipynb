```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread("plate number.jpg")
plt.imshow(img)
plt.show()
```

image preprocessing converting image to RGB, gray, blur, canny edge, dilation, erosion

```python
img1 = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.title("RGB")
plt.imshow(img1)
plt.show()
```

```python
img2 = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.title("gray")
plt.imshow(img2)
plt.show()
```

```python
img3 = cv.GaussianBlur(img, (13, 13), 0)
plt.title('blur')
plt.imshow(img3)
plt.show()
```

```python
img4 = cv.Canny(img3, 100, 200)
plt.title('canny edge')
plt.imshow(img4)
plt.show()
```
